```python
import requests
import bs4

import pandas as pd
```

```python
pages = []
prices = []
ratings = []
titles = []
urls = []

noOfPage = 1

for i in range(1,noOfPage+1):
    pages.append(f"https://books.toscrape.com/catalogue/page-{i}.html")

pages
```

```python
for item in pages:
    page = requests.get(item)
    soup = bs4.BeautifulSoup(page.text, "html.parser")

print(soup)
```

```python
print(soup.prettify())
```

```python
allTitles = soup.find_all("a")
for title in allTitles: 
    if "title" in title.attrs:
        titles.append(title["title"])
    else:
        continue
```

```python
titles
```

```python
for price in soup.find_all("p", class_="price_color"):
    prices.append(price.text)
```

```python
prices
```

```python
for star in soup.find_all("p", class_="star-rating"):
    for k,v in star.attrs.items():
        ratings.append(v[1])
```

```python
ratings
```

```python
for url in soup.find_all("img"):
    url = "http://books.toscrape.com/" + url["src"]
    url = url.replace("..","")
    urls.append(url)
```

```python
urls
```

```python
webData = {"Title":titles, "Price":prices, "Ratings":ratings, "URL":urls}

df = pd.DataFrame(webData)
df
```

```python
df.index +=1
```

```python
df.head()
```

```python
df["Price"] = df["Price"].str.replace("Â£"," ")
df.head()
```

```python
df.sort_values(by="Price", ascending=False)
```

```python
df["Ratings"] = df["Ratings"].replace({"One":1,"Two":2,"Three":3,"Four":4,"Five":5})
df.head()
```

```python
df["Price"] = df["Price"].astype(float)
df.dtypes
```
